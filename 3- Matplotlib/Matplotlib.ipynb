# Introducción Matplotlib

Matplotlib es una librería de python que nos permite realizar gráficos en dos dimensiones para utilizarlos en distintas aplicaciones que van desde notebooks de jupyther hasta páginas web o scripts.  

Lo que matplotlib intenta es hacer sencillas las cosas sencillas y posibles las cosas difíciles. Al usuario principiante le permite realizar gráficas sencillas con pocas línes de código, mientras que al avanzado le permite tener control de caracteristicas como el tipo de línea, colores, anotaciones, etc.  

Por convención matplotlib se importa de la siguiente manera:  

```python
import matplotlib.pyplot as plt
```

Para este notebook también importaremos numpy.

```python
import numpy as np
```

# Ejemplos de gráficos
-----
A continuación, se muestran algunos de los gráficos más comunes que podemos realizar con numpy.

## Gráfico de línea

```python
x = np.linspace(0, 10)
y = np.cos(x)

plt.plot(x, y)
plt.show()
```

## Gráfico de dispersión

```python
x = np.linspace(0, 10)
y = x**2 + 5*x + ( np.random.randn(50) * 20)

plt.scatter(x, y)
plt.show()
```

## Gráfico de barras

```python
etiquetas = ['a', 'b', 'c']
valores = [1, 2, 3]

plt.bar(etiquetas, valores)
plt.show()
```

## Histograma

```python
valores = np.random.randn(1000)
```

```python
plt.hist(valores)
plt.show()
```

## Gráfico de pastel

```python
etiquetas = ['a', 'b', 'c']
valores = [1, 2, 3]

plt.pie(valores, labels=etiquetas)
plt.show()
```

# Figuras y ejes

Existen dos tipos de objeto que matplotlib utiliza para generar gráficos.  

***Figura:*** Es el contenedor principal para todos los elementos del gráfico.  
***Eje:*** Es un área dentro de la figura que contiene el gráfico.  

Estos dos objetos son creados en automático cuando generamos una gráfica, pero tambien podemos crearlos antes de graficar para tener un mejor control en la distribución del resultado.  



## plt.subplots()
Este método genera una figura y un conjunto de ejes deacuerdo al número de filas y columnas que se soliciten.

```python
fig, ax = plt.subplots(2, 2)

ax[0,0].text(.3, .4, 'ax[0,0]', size=20)
ax[0,1].text(.3, .4, 'ax[0,1]', size=20)
ax[1,0].text(.3, .4, 'ax[1,0]', size=20)
ax[1,1].text(.3, .4, 'ax[1,1]', size=20)

plt.show()
```

```python
etiquetas = ['a', 'b', 'c']
valores = [1, 2, 3]

fig, ax = plt.subplots(1, 2, figsize=(12,6))

ax[0].bar(etiquetas, valores)
ax[0].set_title('Grafico de barras')

ax[1].pie(valores, labels=etiquetas)
ax[1].set_title('Grafico de pastel')

plt.show()
```
