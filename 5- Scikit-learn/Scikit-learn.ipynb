# Scikit-learn

Scikit-lear es una librería de aprendizaje automático de python.  

Incluye algoritmos que nos sirven para solucionar distintos tipos de problemas. Además incluye herramientas que nos facilitan el trabajo en distintas etapas de la creación de un modelo.

## API

Una de las caracteristicas de Scikit-learn es su API. Esta se distingue por ser muy limpia y uniforme.  

Scikit-learn incluye clases para cada uno de sus algoritmos y todas ellas funcionan de manera muy similar, por lo que al aprender a usar uno es relativamente sencillo aprender a usar un algoritmo similar.

## Aprendizaje supervisado

A continuación se mencionan algunos de los algoritmos disponibles para aprendizaje supervisado. Para una lista más completa puedes revisar la [documentación](https://scikit-learn.org/stable/supervised_learning.html)

### Regresion

Algunos de los algoritmos disponibles son:

* Regresión lineal
* Árboles de decisión
* Decenso estocástico en gradiente  

### Clasificación

* Regresión logística
* Árboles de decisión
* Máquinas de soporte vectorial (SVM)
* red neuronal simple de perceptrones

### Ejemplo

```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

***Generamos datos para el ejemplo***

```python
np.random.seed(42)

x = np.arange(1000).reshape(-1,1)
y = 3*x + 1200 + (np.random.randn(1000, 1) * 400)

plt.scatter(x, y, s=1)
plt.show()
```

***Modelo***

```python
# Conjuntos de entrenamiento y prueba
x_train, x_test, y_train, y_test = train_test_split(x, y)
```

```python
# Creación y entrenamiento del estimador
estimador = LinearRegression()
estimador.fit(x_train, y_train)

# Predicción
y_pred = estimador.predict(x_test)
```

***Resultado***

```python
# Ecuación de la linea
m = estimador.coef_[0,0]
b = estimador.intercept_[0]
print('La línea que encontramos con este modelos es la siguiente:\n')
print('y = {:.2f} x + {:.2f}\n'.format(m, b))

# Muestra el resultado
plt.scatter(x_test, y_test, s=1)
plt.scatter(x_test, y_pred, s=1, c='red')
plt.show()

# Metrica del error
rmse = mean_squared_error(y_test, y_pred) ** .5
print('\n\nRMSE: {:.2f}\n'.format(rmse))
```

## Aprendizaje No-Supervisado

Scikit-learn tambin incluye algoritmos para aprendisaje no-supervisado que nos ayudan con problemas como:

* Clustering
* Detección de anomalias
* Reducción de dimensionalidad

Para una lista más detallada puedes revisar la [documentación](https://scikit-learn.org/stable/unsupervised_learning.html)

### Ejemplo

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

***Generamos datos para el ejemplo***

```python
features, _ = make_blobs(1000, cluster_std=3, random_state=42)

plt.scatter(features[:,0], features[:,1], s=5)
```

```python
# Crea el modelo
clasificador = KMeans(3)

# Entrena y genera resultado
clusters = clasificador.fit_predict(features)
```

***Resultado***

```python
# Muestra el resultado
plt.scatter(features[:,0], features[:,1], s=5, c=clusters)
plt.show()

# Metrica
print('\nInercia: {:,.2f} \n'.format(clasificador.inertia_))
```
